# # custom infos로 변경
---

```python
import pickle as pkl
```

```python
with open('/content/drive/MyDrive/workspace/study/3D_Object_Detection/OpenPCDet/data/kitti/kitti_infos_val.pkl', 'rb') as f:
    val = pkl.load(f)
```

```python
import numpy as np

new_vals = []
for i in np.arange(len(val)):
    point_cloud = val[i]['point_cloud']
    name = val[i]['annos']['name']
    name = np.where((name!='Car') & (name!='Pedestrian') & (name!='Cyclist') & (name!='DontCare'), 'Cyclist', name)
    # name = np.where((name!='Car') & (name!='Pedestrian'), 'Cyclist', name)
    gt_boxes_lidar = val[i]['annos']['gt_boxes_lidar']
    new_val = {'point_cloud': point_cloud,
               'annos': {'name': name,
                         'gt_boxes_lidar': gt_boxes_lidar}}
    new_vals.append(new_val)
```

```python
class_ = []
for i in new_vals:
    for j in i['annos']['name']:
        class_.append(j)

set(class_)
```

```python
new_val
```

```python
with open('/content/drive/MyDrive/workspace/study/3D_Object_Detection/OpenPCDet/data/kitti/custom_infos_val.pkl', 'wb') as f:
    pkl.dump(new_vals, f)
```

# # bin to npy
---

```python
import numpy as np
import os
import glob

src = '/content/drive/MyDrive/workspace/study/3D_Object_Detection/OpenPCDet/data/kitti/training/velodyne'
velodynes = sorted(os.listdir(src))

for velodyne in velodynes:
    xyzi = np.fromfile(f'{src}/{velodyne}', dtype=np.float32).reshape(-1, 4)
    name = velodyne[:6]
    np.save(f'/content/drive/MyDrive/workspace/study/3D_Object_Detection/OpenPCDet/data/kitti/points/{name}.npy', xyzi)
```
